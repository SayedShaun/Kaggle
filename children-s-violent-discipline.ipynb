```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/kaggle/input/child-abuse-rights/childrens_violence new.csv")
df.head()
```

```python
df.rename(columns={df.columns[-1]:"Childrens Violent Discipline Rate"}, inplace=True)
df.head()
```

```python
df.shape
```

```python
def distplot(column):
    plt.figure(figsize=(8, 3))
    x = df[column]
    sns.distplot(x)
    plt.show()
```

```python
distplot("Childrens Violent Discipline Rate")
```

```python
distplot("Year")
```

```python
def boxplot(column):
    plt.figure(figsize=(8, 3))
    x = df[column]
    sns.boxplot(x=x)
    plt.show()
```

```python
boxplot("Childrens Violent Discipline Rate")
```

```python
boxplot("Year")
```

```python
plt.figure(figsize=(7, 3))
top_10_countries = df.sort_values(by=df.columns[-1], ascending=False).head(10)
sns.countplot(data=top_10_countries, y="Entity", hue="Childrens Violent Discipline Rate")
plt.title("Top 10 Countries with Most Childrens Violent Discipline Rate")
plt.show()
```

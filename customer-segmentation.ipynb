```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/kaggle/input/customer-segmentation-data/customer_segmentation_data.csv")
df.head()
```

```python
sns.pairplot(data=df, hue="Segmentation Group")
```

```python
def category_plot(column: str)->None:
    df[column].value_counts().plot(kind="bar")
```

```python
category_plot("Segmentation Group")
```

```python
category_plot("Gender")
```

```python
category_plot("Geographic Information")
```

```python
category_plot("Preferred Communication Channel")
```

```python
category_plot("Preferred Language")
```

```python
def boxplot(column: str)->None:
    sns.boxenplot(data=df, x=column)
```

```python
boxplot("Age")
```

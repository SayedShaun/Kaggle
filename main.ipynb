```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_csv("train.csv")
df
```

```python
y = df["Transported"]
df.drop(["Transported", "PassengerId"], axis=1, inplace=True)
df
```

```python

plt.figure(figsize=[20, 10])
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder
def label_encoder(data):
    le = LabelEncoder()
    for i in data.columns:
        if data[i].dtype == "object" or data[i].dtype =="bool":
            data[i] = le.fit_transform(data[i])

    return data

label_encoder(df)
df
```

```python
def handle_missing(data):
    for i in data.columns:
        if data[i].isnull().any():
            data[i].fillna(np.mean(data[i]), inplace = True)

    return data

handle_missing(df)

df.isnull().sum()
```

```python
df_x = df.select_dtypes(exclude="object")
x = df_x.drop(["Name"], axis=1)
x.isnull().sum()
```

```python

import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units = 32, activation = "relu"),
    tf.keras.layers.Dense(units = 16, activation = "relu"),
    tf.keras.layers.Dense(units = 8, activation = "relu"),
    tf.keras.layers.Dense(units = 1, activation = "sigmoid")
])


model.compile(optimizer= tf.keras.optimizers.Adam(learning_rate=0.001) ,loss = tf.keras.losses.BinaryCrossentropy())
model.fit(x, y,validation_data=(x, y), epochs=100)
```

```python
test_df = pd.read_csv("test.csv")
pass_id = test_df["PassengerId"]
test_df.drop(["PassengerId","Name"], axis=1, inplace=True)
```

```python
x_test = label_encoder(test_df)
x_test = handle_missing(test_df)
x_test
```

```python
t = model.predict(x_test)
t
```

```python
arr = np.array([True if i > 0.5  else False for i in t])

arr_final = pd.DataFrame(arr)
pass_id = pd.DataFrame(pass_id)
pass_id["Transported"] = arr

pass_id.to_csv(r"C:\Users\User\Desktop\Result\sample_submission.csv", index=False)
```

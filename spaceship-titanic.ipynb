```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
df
```

```python
y = df["Transported"]
df.drop(["Transported", "PassengerId"], axis=1, inplace=True)
```

```python

plt.figure(figsize=[20, 10])
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
def delete_object(data):
    for i in data.columns:
        if data[i].dtype == "object":
            data.drop([i], axis = 1, inplace = True)

    return data

delete_object(df)
```

```python
def handle_missing(data):
    for i in data.columns:
        if data[i].isnull().any():
            data[i].fillna(np.mean(data[i]), inplace = True)

    return data

handle_missing(df)

df
```

```python
x = df
x
```

```python
#plt.figure(figsize=[20, 10])
#sns.pairplot(x)
```

Using Tensorflow.Keras

```python

import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units = 64, activation = "relu", kernel_regularizer=tf.keras.regularizers.l2(0.1)),
    tf.keras.layers.Dense(units = 32, activation = "relu", kernel_regularizer=tf.keras.regularizers.l2(0.1)),
    tf.keras.layers.Dense(units = 16, activation = "relu", kernel_regularizer=tf.keras.regularizers.l2(0.1)),
    tf.keras.layers.Dense(units = 1, activation = "sigmoid")
])


model.compile(optimizer= tf.keras.optimizers.Adam(learning_rate=0.001) ,loss = tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
model.fit(x, y, validation_data=(x, y),  epochs=100, batch_size=32)
```

```python
test_df = pd.read_csv("/kaggle/input/spaceship-titanic/test.csv")
pass_id = test_df["PassengerId"]
test_df.drop(["PassengerId","Name"], axis=1, inplace=True)
```

```python
x_test = delete_object(test_df)
x_test = handle_missing(test_df)
x_test
```

```python
t = model.predict(x_test)
t
```

```python
arr = np.array([True if i > 0.5  else False for i in t])

arr_final = pd.DataFrame(arr)
pass_id = pd.DataFrame(pass_id)
pass_id["Transported"] = arr

#pass_id.to_csv(r"C:\Users\User\Desktop\Result\sample_submission.csv", index=False)
```

```python
# Assuming you have trained the model using model.fit() before this point

# Evaluate the model on the training data (you can also use a separate validation set here)
loss, accuracy = model.evaluate(x, y)

print("Loss:", loss)
print("Accuracy:", accuracy)
```
